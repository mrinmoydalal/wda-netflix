# Importing

```python
import pandas as pd

#For NLP
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

#For sentiment analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
nltk.download('vader_lexicon')
```

# Fetching the scraped reviews

```python
df = pd.read_csv('RottenTomatoesReviews.csv')
```

```python
df
```

# Tokenizing 

```python
tokenizedReviews = []
for review in df['Review']:
    tokenizedReviews.append(nltk.word_tokenize(str(review).lower()))
```

```python
len(tokenizedReviews)
```

```python
tokenizedReviews[:10]
```

# Lemmatization

```python
lemmatizedWords = [[WordNetLemmatizer().lemmatize(word) for word in reviewToken] for reviewToken in tokenizedReviews]
lemmatizedWords[5:]
```

```python
stopwordsforremoval = [[token for token in sublist if not token in stopwords.words('english') if token.isalpha()] for sublist in lemmatizedWords]
```

# Removing stop words

```python
stopwordsforremoval[:10]
```

# Combining the words

```python
combinedWords = []
for word in stopwordsforremoval:
    combinedWords.append(" ".join(word))
combinedWords[:10]
```

# Calculating the sentiment scores

```python
negReviewScore = []
posReviewScore = []
neuReviewScore = []
compoundedReviewScore = []

for sentence in combinedWords:
    
    pol = sid.polarity_scores(sentence)
    negReviewScore.append(pol['neg'])
    neuReviewScore.append(pol['neu'])
    posReviewScore.append(pol['pos'])
    compoundedReviewScore.append(pol['compound'])
```

# Saving the output in a csv file

```python
df['Compounded Review Score'] = compoundedReviewScore 
df['Positive Review Score'] = posReviewScore
df['Negative Review Score'] = negReviewScore
df['Neutral Review Score'] = neuReviewScore

df.to_csv('Score.csv') 
```
